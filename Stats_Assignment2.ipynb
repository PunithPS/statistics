```python
#* Continuous Random Variable Distribution:
# Q3. Given the data on the amount of rainfall (in mm) over 20 days: [3.2, 4.1, 2.8, 3.6, 3.3, 4.0, 2.9, 3.5, 3.1, 4.2, 2.7, 3.4, 3.0, 3.9, 2.6, 3.7, 3.2, 4.1, 2.8, 3.6],
#calculate the mean and standard deviation. Plot the histogram and fit a normal distribution to this data.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Data on the amount of rainfall in mm
rainfall_data = [3.2, 4.1, 2.8, 3.6, 3.3, 4.0, 2.9, 3.5, 3.1, 4.2,
                 2.7, 3.4, 3.0, 3.9, 2.6, 3.7, 3.2, 4.1, 2.8, 3.6]

# Calculateing mean and standard deviation
mean = np.mean(rainfall_data)
std_dev = np.std(rainfall_data)

print(f"Mean of rainfall data: {mean:.2f} mm")
print(f"Standard deviation of rainfall data: {std_dev:.2f} mm")

# Ploting the  histogram
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(rainfall_data, bins=10, density=True, alpha=0.6, color='g', edgecolor='black')

# Fit a normal distribution
mu, std = mean, std_dev
pdf = stats.norm.pdf(bins, mu, std)
plt.plot(bins, pdf, 'k', linewidth=2)

# Title and labels
plt.title('Histogram of Rainfall with Normal Distribution Fit')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Density')

# Show plot
plt.show()

```

```python
# Probability of Continuous Variable:
#Q4. Using the rainfall data from the previous question, calculate the probability that the rainfall on a given day will be between 3.0 mm and 3.5 mm, assuming a normal distribution.
import numpy as np
import scipy.stats as stats

# Data on the amount of rainfall in mm
rainfall_data = [3.2, 4.1, 2.8, 3.6, 3.3, 4.0, 2.9, 3.5, 3.1, 4.2,
                 2.7, 3.4, 3.0, 3.9, 2.6, 3.7, 3.2, 4.1, 2.8, 3.6]

# Calculate mean and standard deviation
mean = np.mean(rainfall_data)
std_dev = np.std(rainfall_data, ddof=1)  # ddof=1 for sample standard deviation

# Define the range
lower_bound = 3.0
upper_bound = 3.5

# Calculate the CDF values for the bounds
prob_lower = stats.norm.cdf(lower_bound, mean, std_dev)
prob_upper = stats.norm.cdf(upper_bound, mean, std_dev)

# Probability that rainfall is between 3.0 mm and 3.5 mm
probability = prob_upper - prob_lower

print(f"Mean of rainfall data: {mean:.2f} mm")
print(f"Standard deviation of rainfall data: {std_dev:.2f} mm")
print(f"Probability that rainfall is between {lower_bound} mm and {upper_bound} mm: {probability:.4f}")
```

```python
#Q6. Given a normally distributed dataset representing the claim amounts with a mean of $10,000 and a standard deviation of $2,000,
#calculate the Z-score for a claim amount of $12,000. Then, determine the probability that a randomly selected claim will exceed $12,000.

import scipy.stats as stats

# Given values
mean = 10000
std_dev = 2000
claim_amount = 12000

# Calculate the Z-score
z_score = (claim_amount - mean) / std_dev

# Calculate the cumulative probability for the Z-score
cumulative_prob = stats.norm.cdf(z_score)

# Calculate the probability of exceeding the claim amount
prob_exceed = 1 - cumulative_prob

# Print results
print(f"Z-score for a claim amount of ${claim_amount}: {z_score:.2f}")
print(f"Probability that a claim will exceed ${claim_amount}: {prob_exceed:.4f}")
```

```python
# * Application of 68-95-99.7 Rule:
#Q7. In a normally distributed dataset with a mean of 75 and a standard deviation of 10,
#calculate the range of values that fall within one, two, and three standard deviations from the mean. How much data falls within these ranges?
import scipy.stats as stats
mean = 75
std_dev = 10

one_stnd_range = (mean - std_dev, mean + std_dev)
two_stnd_range = (mean - 2 * std_dev, mean + 2 * std_dev)
three_stnd_range = (mean - 3 * std_dev, mean + 3 * std_dev)

print(f"Range within one standard deviation: {one_stnd_range}")
print(f"Range within two standard deviations: {two_stnd_range}")
print(f"Range within three standard deviations: {three_stnd_range}")

prob_within_one_std = stats.norm.cdf(mean + std_dev, mean, std_dev) - stats.norm.cdf(mean - std_dev, mean, std_dev)
prob_within_two_std = stats.norm.cdf(mean + 2 * std_dev, mean, std_dev) - stats.norm.cdf(mean - 2 * std_dev, mean, std_dev)
prob_within_three_std = stats.norm.cdf(mean + 3 * std_dev, mean, std_dev) - stats.norm.cdf(mean - 3 * std_dev, mean, std_dev)



print(f"Percentage of data within one standard deviation: {prob_within_one_std * 100:.2f}%")
print(f"Percentage of data within two standard deviations: {prob_within_two_std * 100:.2f}%")
print(f"Percentage of data within three standard deviations: {prob_within_three_std * 100:.2f}%")
```

```python
# Q8. In a study of the heights of a certain plant species, the heights are normally distributed with a mean of 60 cm and
#a standard deviation of 5 cm. Using the 68-95-99.7 rule, determine the range of heights that fay within one, two, and three standard deviations from the mean.
#Then, calculate the probability that a randomly selected plant will have a height between 55 cm and 65 cm.

import scipy.stats as stats

mean = 60
stand_dev = 5

one_stand_range = (mean - std_dev, mean + stand_dev)
two_stand_range = (mean - 2 * std_dev, mean + 2 * stand_dev)
three_stand_range = (mean - 3 * std_dev, mean + 3 * stand_dev)

print(f"Range within one standard deviation: {one_stand_range}")
print(f"Range within two standard deviations: {two_stand_range}")
print(f"Range within three standard deviations: {three_stand_range}")

lower_bound = 55
upper_bound = 65

prob_lower = stats.norm.cdf(lower_bound, mean, std_dev)
prob_upper = stats.norm.cdf(upper_bound, mean, std_dev)

# Probability of height being between 55 cm and 65 cm
probability = prob_upper - prob_lower


print(f"Probability that a plant height is between 55 cm and 65 cm: {probability:.4f}")
```

```python
#* Probability Density Function:
# Q9. For a normal distribution with a mean of 50 and a standard deviation of 5, calculate the value of the probability density function at x=55
import numpy as np

# Given values
mean = 50
std_dev = 5
x = 55

# Calculate the PDF value
pdf_value = (1 / (np.sqrt(2 * np.pi * std_dev**2))) * np.exp(-((x - mean)**2) / (2 * std_dev**2))

# Print the result
print(f"The value of the probability density function at x = {x} is {pdf_value:.4f}")
```

```python
# Q10. * Convert the following data points to Z-scores given a mean of 20 and a standard deviation of 4: [16, 24, 20, 28, 12).
#Plot these Z-scores on a standard normal distribution curve.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


mean = 20
stand_deva = 4
points = [16, 24, 20, 28, 12]
points_to_plot = []

for i in points:
  z = (i - mean)/stand_deva
  points_to_plot.append(z)

print(points_to_plot)

x = np.linspace(-4,4,1000)

pdf = stats.norm.pdf(x,0,1)

plt.plot(x, pdf, label='Standard Normal Distribution', color='blue')

plt.scatter(points_to_plot, stats.norm.pdf(points_to_plot, 0, 1), color='red', zorder=5, label='Z-scores')
for z in points_to_plot:
    plt.text(z, stats.norm.pdf(z, 0, 1), f'{z:.1f}', fontsize=12, ha='right')

# Set labels and title
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.title('Z-scores on Standard Normal Distribution Curve')
plt.legend()


plt.show()
```

```python
#Fitting Normal Distribution:
#q11. Given the data on the amount of rainfall (in mm) over 20 days: [3.2, 4.1, 2.8, 3.6, 3.3, 4.0, 2.9, 3.5, 3.1, 4.2, 2.7, 3.4, 3.0, 3.9, 2.6, 3.7, 3.2, 4.1, 2.8, 3.6],
#calculate the parameters (mean and standard deviation) for the fitted normal distribution.
#Use these parameters to estimate the probability of rainfall exceeding 4.0 mm on a given day.

import numpy as np
import scipy.stats as stats

# Rainfall data
rainfall_data = np.array([3.2, 4.1, 2.8, 3.6, 3.3, 4.0, 2.9, 3.5, 3.1, 4.2, 2.7, 3.4, 3.0, 3.9, 2.6, 3.7, 3.2, 4.1, 2.8, 3.6])


mean = np.mean(rainfall_data)
std_dev = np.std(rainfall_data)

# Calculate the probability of rainfall exceeding 4.0 mm
value = 4.0
prob_exceeding = 1 - stats.norm.cdf(value, mean, std_dev)

# Print results
print(f"Mean of the rainfall data: {mean:.2f} mm")
print(f"Standard deviation of the rainfall data: {std_dev:.2f} mm")
print(f"Probability of rainfall exceeding {value} mm: {prob_exceeding:.4f}")
```

```python
# Q12. You have the following data on monthly rainfall (in mm) over a year: [110, 105, 120, 95, 130, 115, 100, 125, 110, 105, 130, 120].
#Plot the histogram for this data and fit a normal distribution. Calculate the mean and standard deviation of the fitted normal distribution and interpret the result.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given data
rainfall_data = np.array([110, 105, 120, 95, 130, 115, 100, 125, 110, 105, 130, 120])

# Calculate mean and standard deviation
mean = np.mean(rainfall_data)
std_dev = np.std(rainfall_data)

# Plot histogram
plt.hist(rainfall_data, bins=8, density=True, alpha=0.6, color='g', edgecolor='black')

# Plot normal distribution fit
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

# Add labels and title
plt.xlabel('Rainfall (mm)')
plt.ylabel('Probability Density')
plt.title('Histogram and Normal Distribution Fit')


# Show plot
plt.show()

# Print mean and standard deviation
print(f"Mean of the rainfall data: {mean:.2f} mm")
print(f"Standard deviation of the rainfall data: {std_dev:.2f} mm")
```
