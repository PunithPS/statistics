```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Baseball.csv')
```

```python
df
```

```python
#getting the mean of 'RS' column
df['RS'].mean()
```

```python
#mean of ARI team for 'RS' column till 2005
df[(df['Team']=='ARI') & (df['Year'] <=2005)]['RS'].mean()
```

```python
#mean of run allowed (RA) by the chicago (CHC) team untill 2007
df[(df['Team']=='CHC') & (df['Year'] <=2007)]['RA'].mean()
```

```python
#what is the median of wins of the baltimore team (BAL) untill 2007

df[(df['Team']=='BAL') & (df['Year'] <=2000)]['W'].median()
```

```python
#mode of "OBP" for Baltimore team till 2003

df[(df['Team']=="BAL") & (df['Year']<= 2003)]["OBP"].mode()[0]
```

```python
#Getting the idea about the correlation   pearsons's correlation co efficient
df[['RS','RA','W']].corr()
```

```python
'''
Pearson's correlation coefficient ranges between [-1,1]
-1 indicates perfect negative correlation, +1 indicates perfect positive correlation
0 indicates that the two variables are alomost uncorrelated
'''
```

```python
#correlation heatmap
import seaborn as sns
df_corr = df[['RS','RA','W']].corr()
sns.heatmap(df_corr, annot = True, cmap = 'coolwarm')
```

```python
#create a heatmap for correlation between OBP, SLG and BA
df_corr1 = df[['OBP','SLG','BA']].corr()
sns.heatmap(df_corr1, annot = True, cmap = 'coolwarm')
```
